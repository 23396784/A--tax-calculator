```python
# converting weekly salary to annull salary
# Annual salary = weekly salary * 52
import math

import math

# Rounded weekly salaries generated in question 1
weeklySalary = [1693, 1358, 1772, 2234, 1308, 1308, 2263, 1835, 1184, 1717]

# Step 1: Convert weekly salaries to annual incomes
annualIncomes = [salary * 52 for salary in weeklySalary]

# Step 2: Define tax calculation function
def calculate_tax(annualIncome):
    if annualIncome <= 18200:
        return 0  # No tax deduction for income <= 18200
    elif annualIncome <= 45000:
        return 0.19 * (annualIncome - 18200)  # 19c for each $1 over $18,200
    elif annualIncome <= 120000:
        return 5092 + 0.325 * (annualIncome - 45000)  # $5,092 + 32.5c for each $1 over $45,000
    elif annualIncome <= 180000:
        return 29467 + 0.37 * (annualIncome - 120000)  # $29,467 + 37c for each $1 over $120,000
    else:
        return 51667 + 0.45 * (annualIncome - 180000)  # $51,667 + 45c for each $1 over $180,000

# Step 3: Calculate income tax for each individual and round up
incomeTax = [math.ceil(calculate_tax(income)) for income in annualIncomes]

# Step 4: Print the results
print("Annual Incomes:", annualIncomes)
print("Rounded-Up Income Tax:", incomeTax)
```

```python
# Identifying the range of weekly salary is of utmost importance to calculate the weekly tax

# Applying the coeficient y = a *(x + 0.99)- b), by adding $0.99 to calculate witholding tax income

# Method: loop through each weekly salary in the weekly salary list
```

```python
# Using the Weekly salaries from Q1
weeklySalary = [1693, 1358, 1772, 2234, 1308, 1308, 2263, 1835, 1184, 1717]

 # Defining the ranges associated with the coefficients (a, b) 
# for an enabling environment to calculate the withholding weekly income tax
ranges = [
    (359, 438, 0.1900, 68.3462),
    (438, 548, 0.2900, 112.1942),
    (548, 721, 0.2100, 68.3465),
    (721, 865, 0.2190, 74.8369),
    (865, 1282, 0.3477, 186.2119),
    (1282, 2307, 0.3450, 182.7504),
    (2307, 3461, 0.3900, 286.5965),
    (3461, float("inf"), 0.4700, 563.5196),
]
# Calculate withholding tax for each salary
withholdingTax = []
for salary in weeklySalary:
    if salary < 359:
        withholdingTax.append(0)  # No tax for weekly salaries less than $359
    else:
        # Find the range that the salary belongs to
        for lower, upper, a, b in ranges:
            if lower <= salary < upper:
                # Apply the formula y = a * (x + 0.99) - b
                tax = a * (salary + 0.99) - b
                withholdingTax.append(round(tax, 2))  # Round to 2 decimal places
                break
# Print the calculated withholding taxes
print("Withholding Tax:", withholdingTax)    
```

```python
# Weekly salaries generated in Q1
weeklySalary = [1693, 1358, 1772, 2234, 1308, 1308, 2263, 1835, 1184, 1717]

# Tax ranges for annual income and corresponding tax rates
annual_ranges = [
    (0, 18200, 0, 0),                 # $0 - $18,200: No tax
    (18201, 45000, 0.19, 18200),      # $18,201 - $45,000: 19c for each $1 over $18,200
    (45001, 120000, 0.325, 45000),    # $45,001 - $120,000: $5,092 + 32.5c for each $1 over $45,000
    (120001, 180000, 0.37, 120000),   # $120,001 - $180,000: $29,467 + 37c for each $1 over $120,000
    (180001, float("inf"), 0.45, 180000)  # Over $180,000: $51,667 + 45c for each $1 over $180,000
]

# Withholding tax ranges and coefficients for weekly tax calculation
ranges = [
    (359, 438, 0.1900, 68.3462),
    (438, 548, 0.2900, 112.1942),
    (548, 721, 0.2100, 68.3465),
    (721, 865, 0.2190, 74.8369),
    (865, 1282, 0.3477, 186.2119),
    (1282, 2307, 0.3450, 182.7504),
    (2307, 3461, 0.3900, 286.5965),
    (3461, float("inf"), 0.4700, 563.5196),
]

# Calculate weekly withholding tax
withholdingTax = []
for salary in weeklySalary:
    if salary < 359:
        withholdingTax.append(0)  # No tax for weekly salaries less than $359
    else:
        for lower, upper, a, b in ranges:
            if lower <= salary < upper:
                tax = a * (salary + 0.99) - b
                withholdingTax.append(tax)
                break

# Calculate tax refunds
taxReturn = []
for i, salary in enumerate(weeklySalary):
    annual_income = salary * 52  # Convert weekly salary to annual income
    annual_withholding_tax = withholdingTax[i] * 52  # Convert weekly withholding tax to annual withholding tax
    
    # Compute the annual tax liability
    tax_liability = 0
    for lower, upper, rate, threshold in annual_ranges:
        if lower <= annual_income <= upper:
            tax_liability += rate * (annual_income - threshold)
            break
        elif annual_income > upper:
            tax_liability += rate * (upper - threshold)
    
    # Calculate the refund
    refund = annual_withholding_tax - tax_liability
    taxReturn.append(round(refund, 2))

# Print the calculated tax refunds
print("Tax Refunds for Each Individual:", taxReturn)
```

```python
For the 11% superannuation contribution: 
I need to adjust Tax Calculations because of the 11% superannuation contribution
I have to exclude the 11% superannuation contribution from The weekly taxable salary before calculating withholding tax and the annual taxable salary
used to calculate income hence, Include Superannuation in the Output.
For the purpose of clarity i include the superannuation contribution amount for each individual in the output 

Revised Calculation Method
Superannuation Contribution = Basic Weekly Salary × 0.11 
To Calculate the taxable salary:
Taxable Weekly Salary=Base Weekly Salary−Superannuation Contribution
Use the taxable salary for withholding tax calculations and annual income tax calculations.
Finally, we Print the superannuation contribution as part of the output.
```

```python
for i, salary in enumerate(weeklySalary):
    # Calculate superannuation contribution
    superannuation = salary * 0.11  # 11% superannuation contribution

    # Calculate taxable salary
    taxable_salary = salary - superannuation

    # Calculate annual income
    annual_income = taxable_salary * 52  # Convert weekly taxable salary to annual income
    annual_withholding_tax = withholdingTax[i] * 52  # Convert weekly withholding tax to annual withholding tax

    # Compute the annual tax liability
    tax_liability = 0
    for lower, upper, rate, threshold in ranges:
        if lower <= salary < upper:
            tax = rate * (salary + 0.99) - threshold
            withholdingTax.append(tax)
            break  # Exit loop after correct range is found
        elif annual_income > upper:
            tax_liability += rate * (upper - threshold)

    # Calculate the refund
    refund = annual_withholding_tax - tax_liability

    # Calculate weekly income after withholding tax
    weekly_income_after_tax = salary - withholdingTax[i]

    # Print details
    print(f"## Person {i + 1}")
    print(f"Base Weekly Salary:          ${salary:,.2f}")
    print(f"Superannuation Contribution: ${superannuation:,.2f}")
    print(f"Taxable Weekly Salary:       ${taxable_salary:,.2f}")
    print(f"Weekly Withholding Tax:      ${withholdingTax[i]:,.2f}")
    print(f"Weekly Income After Tax:     ${weekly_income_after_tax:,.2f}")
    print(f"Annual Income Tax:           ${tax_liability:,.2f}")
    print(f"Tax Return:                  ${refund:,.2f}\n")
```

```python
# Initial weekly salaries (including super)
weekly_salary = [1693, 1358, 1772, 2234, 1308, 1308, 2263, 1835, 1184, 1717]
SUPER_RATE = 0.11

def calculate_base_and_super(total_salary):
   """Calculate base salary and super from total package"""
   base_salary = total_salary / (1 + SUPER_RATE)
   super_amount = base_salary * SUPER_RATE
   return round(base_salary, 2), round(super_amount, 2)

def calculate_income_tax(annual_income):
   """Calculate annual income tax"""
   if annual_income <= 18200:
       return 0
   elif annual_income <= 45000:
       return 0.19 * (annual_income - 18200)
   elif annual_income <= 120000:
       return 5092 + 0.325 * (annual_income - 45000)
   elif annual_income <= 180000:
       return 29467 + 0.37 * (annual_income - 120000)
   else:
       return 51667 + 0.45 * (annual_income - 180000)

def calculate_withholding_tax(weekly_salary):
   """Calculate weekly withholding tax"""
   ranges = [
       (359, 438, 0.1900, 68.3462),
       (438, 548, 0.2900, 112.1942),
       (548, 721, 0.2100, 68.3465),
       (721, 865, 0.2190, 74.8369),
       (865, 1282, 0.3477, 186.2119),
       (1282, 2307, 0.3450, 182.7504),
       (2307, 3461, 0.3900, 286.5965),
       (3461, float('inf'), 0.4700, 563.5196)
   ]
   
   if weekly_salary < 359:
       return 0
       
   for lower, upper, a, b in ranges:
       if lower <= weekly_salary < upper:
           return round(a * (weekly_salary + 0.99) - b, 2)
   return 0

# Calculate and print results for each person
for i, total_weekly in enumerate(weekly_salary, 1):
   # Calculate base salary and super
   base_weekly, super_weekly = calculate_base_and_super(total_weekly)
   
   # Calculate weekly withholding tax on base salary
   withholding = calculate_withholding_tax(base_weekly)
   
   # Calculate weekly income (base minus withholding and super)
   weekly_income = base_weekly - withholding
   
   # Calculate annual tax amounts
   annual_base = base_weekly * 52
   annual_income_tax = calculate_income_tax(annual_base)
   annual_withholding = withholding * 52
   tax_return = annual_withholding - annual_income_tax
   
   # Print formatted output
   print(f"\n## Person {i} "
         f"weekly salary ${total_weekly:.2f} "
         f"weekly superannuation contribution ${super_weekly:.2f} "
         f"weekly withholding tax ${withholding:.2f} "
         f"weekly income ${weekly_income:.2f} "
         f"income tax ${annual_income_tax:.2f} "
         f"tax return ${tax_return:.2f}.")
```
